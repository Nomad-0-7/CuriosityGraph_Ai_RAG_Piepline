```python
import arxiv
from urllib.request import urlretrieve
import pandas as pd
import os
import re

client = arxiv.Client()

search = arxiv.Search(
    query = 'AI Engineering',
    max_results = 20,
    sort_by = arxiv.SortCriterion.SubmittedDate,
    sort_order = arxiv.SortOrder.Descending
)
results = client.results(search)

data = [
      {
          "title": r.title,
          "pdf_url": r.pdf_url,
          "published": r.published
      }
      for r in client.results(search)
]
df = pd.DataFrame(data)

pd.set_option('display.max_colwidth', None)

print(df["title"])

download_dir = "./knowledge_base"
os.makedirs(download_dir, exist_ok=True)

print("--- Starting Downloads ---")
for index, row in df.iterrows():
    title = row["title"]
    url = row["pdf_url"]
    clean_title = re.sub(r'[^a-zA-Z0-9\s_-]', '', title).strip()
    clean_title = re.sub(r'\s+', ' ', clean_title)
    filename = f"{clean_title}.pdf"
    file_path = os.path.join(download_dir, filename)
    print(f"[{index + 1}/{len(df)}] Downloading: {title}")
    try:
        urlretrieve(url, file_path)
        print(f"✓ Saved to: {file_path}\n")
    except Exception as e:
        print(f"✗ Failed to download: {e}\n")
print("--- Data Ingestion Complete ---")
```

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Natively reads all PDFs in a directory using pure python pypdf
loader = PyPDFDirectoryLoader("./knowledge_base")

docs = loader.load()
print(f"Loaded {len(docs)} pages from your PDFs.")

# Chunk your extracted clean text
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(docs)
print(f"Split documents into {len(chunks)} chunks.")
```
